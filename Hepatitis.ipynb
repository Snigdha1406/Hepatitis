Centrality Measures

```python
print('Mean protime : {0}'.format(data.protime.mean()))         # mean
print('Median protime : {0}'.format(data.protime.median()))     # median
```

Dispersion measures

```python
print('protime range : {0}'.format(data.protime.max()  - data.protime.min())) # range
print('Variance protime  : {0}'.format(data.protime.var()))  #var
print('Standard deviation protime : {0}'.format(data.protime.std())) # standard deviation
```
